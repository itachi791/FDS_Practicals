```python
import math
```

```python
# Function to calculate nCr (combinations)
def nCr(n, r):
    if r < 0 or r > n:
        return 0
    return math.factorial(n) // (math.factorial(r) * math.factorial(n - r))

# Binomial PMF: P(X = x)
def binomial_pmf(n, p, x):
    q = 1 - p
    return nCr(n, x) * (p ** x) * (q ** (n - x))

# P(X ≤ x)
def binomial_atmost(n, p, x):
    total = 0
    for k in range(0, x + 1):
        total += binomial_pmf(n, p, k)
    return total

# P(X ≥ x)
def binomial_atleast(n, p, x):
    if x == 0:
        return 1
    return 1 - binomial_atmost(n, p, x - 1)


print("=== Binomial Distribution Calculator ===")
print("1. Probability of EXACT X")
print("2. Probability of AT MOST X")
print("3. Probability of AT LEAST X")


choice = int(input("Enter your choice (1/2/3): "))
n = int(input("Enter n (number of trials): "))
p = float(input("Enter p (success probability): "))
x = int(input("Enter X value: "))

#Chck if X is within the valid range
if x < 0 or x > n:
    print("Invalid value for X. X must be between 0 and n (inclusive).")
else:
    if choice == 1:
        ans = binomial_pmf(n, p, x)
        print(f"\nP(X = {x}) = {ans:.6f}")

    elif choice == 2:
        ans = binomial_atmost(n, p, x)
        print(f"\nP(X ≤ {x}) = {ans:.6f}")
 
    elif choice == 3:
        ans = binomial_atleast(n, p, x)
        print(f"\nP(X ≥ {x}) = {ans:.6f}")

    else:
        print("Invalid choice!")
```
