```python
import pandas as pd
import math

df = pd.read_csv(r"D:\College\FDS Experiments\archive\iris.csv")

x = df['SepalLengthCm'].tolist()
y = df['PetalLengthCm'].tolist()
n = len(x)

#MEAN
mean_x = sum(x) / n
mean_y = sum(y) / n
print("Mean of Sepal Length:", mean_x)
print("Mean of Petal Length:", mean_y)
print()

#MEDIAN
def get_median(values):
    sorted_vals = sorted(values)
    m = len(sorted_vals)
    if m % 2 == 0:
        return (sorted_vals[m//2 - 1] + sorted_vals[m//2]) / 2
    else:
        return sorted_vals[m//2]

median_x = get_median(x)
median_y = get_median(y)
print("Median of Sepal Length:", median_x)
print("Median of Petal Length:", median_y)
print()

#MODE
def get_mode(values):
    freq = {}
    for v in values:
        freq[v] = freq.get(v, 0) + 1
    max_freq = max(freq.values())
    modes = [k for k, v in freq.items() if v == max_freq]
    return modes

mode_x = get_mode(x)
mode_y = get_mode(y)
print("Mode of Sepal Length:", mode_x)
print("Mode of Petal Length:", mode_y)
print()

#VARIANCE
var_x = sum([(i - mean_x)**2 for i in x]) / (n - 1)
var_y = sum([(i - mean_y)**2 for i in y]) / (n - 1)
print("Variance of Sepal Length:", var_x)
print("Variance of Petal Length:", var_y)
print()

#SD
sd_x = math.sqrt(var_x)
sd_y = math.sqrt(var_y)
print("Standard Deviation of Sepal Length:", sd_x)
print("Standard Deviation of Petal Length:", sd_y)
print()

#COVARIANCE
cov_xy = sum([(x[i] - mean_x) * (y[i] - mean_y) for i in range(n)]) / (n - 1)
print("Covariance between Sepal Length and Petal Length:", cov_xy)
print()

#PEARSON CORRELATION
r_xy = cov_xy / (sd_x * sd_y)
print("Pearson Correlation:", r_xy)
print()

#SKEWNESS
skew_x = (sum([(i - mean_x)**3 for i in x]) / n) / (sd_x**3)
skew_y = (sum([(i - mean_y)**3 for i in y]) / n) / (sd_y**3)
print("Skewness of Sepal Length:", skew_x)
print("Skewness of Petal Length:", skew_y)
print()

#KURTOSIS
kurt_x = ((sum([(i - mean_x)**4 for i in x]) / n) / (sd_x**4)) - 3
kurt_y = ((sum([(i - mean_y)**4 for i in y]) / n) / (sd_y**4)) - 3
print("Kurtosis of Sepal Length:", kurt_x)
print("Kurtosis of Petal Length:", kurt_y)
```
