```python
G = [0,0,0,0,1,1,1,1]
x = [10,12,13,9,11,14,10,13]
y = [15,17,16,14,18,20,17,19]
```

```python
#X0Mean, X1 Mean, Y0 Mean, Y1 Mean
totalX0 = sum(x[0:4])
averagex0 = totalX0 / 4
print("X0 =", averagex0)

totalX1 = sum(x[4:8])
averagex1 = totalX1 / 4
print("X1 =", averagex1)
```

```python
totalY0 = sum(y[0:4])
averagey0 = totalY0 / 4
print("Y0 =", averagey0)

totalY1 = sum(y[4:8])
averagey1 = totalY1 / 4
print("Y1 =", averagey1)
```

```python
#Variability
x_mean = sum(x) / len(x)

y_mean = sum(y) / len(y)

print('xmean = ', x_mean)
print('ymean = ',y_mean)
```

```python
x_minus_x_mean = [i - x_mean for i in x]
x_minus_xmean_sq2 = [(i - x_mean)**2 for i in x]

total_x_minus_xmean_sq2 = sum(x_minus_xmean_sq2)

print("(xi - xmean) =", x_minus_x_mean)
print("(xi - xmean)^2 =", x_minus_xmean_sq2)
print("Sum =", total_x_minus_xmean_sq2)
```

```python
y_minus_y_mean = [i - y_mean for i in y]
y_minus_ymean_sq2 = [(i - y_mean)**2 for i in y]

total_y_minus_ymean_sq2 = sum(y_minus_ymean_sq2)

print("(yi - ymean) =", y_minus_y_mean)
print("(yi - ymean)^2 =", y_minus_ymean_sq2)
print("Sum =", total_y_minus_ymean_sq2)
```

```python
#(xi-xmean)(yi-ymean)
product = [(xi - x_mean) * (yi - y_mean) for (xi, yi) in zip(x, y)]

print("Products =", product)
sum_product = sum(product)
print("Sum of products =", sum_product)
```

```python
#Calculating β₀, βG, and βX
import numpy as np
X = np.column_stack((np.ones(len(G)), G, x))

print("Design Matrix X =\n", X)

# ---------------------------
# 2. Compute Beta = (X'X)^(-1) X'y
# ---------------------------

Xt = X.T
XtX = Xt @ X
XtX_inv = np.linalg.inv(XtX)
Xty = Xt @ y
beta = XtX_inv @ Xty

print("\n(X'X):\n", XtX)
print("\nInverse (X'X)^(-1):\n", XtX_inv)
print("\nX'y:\n", Xty)
print("\nBeta coefficients [β0, βG, βX]:\n", beta)

# Naming them clearly
beta0 = beta[0]
betaG = beta[1]
betaX = beta[2]

print(f"\nβ0  = {beta0:.2f}")
print(f"βG  = {betaG:.2f}")
print(f"βX  = {betaX:.2f}")
```

```python
#Variability of ANCOVA vs Variability of Covariance only Model


#Variability for reduced covariance only model
# g = β0 + βX x X

Sxx = total_x_minus_xmean_sq2
Sxy = sum_product
βx = Sxy/Sxx
βo = y_mean - βx * x_mean
print("Sum of squares for X:",Sxx)
print("Sum of Cross-Products of X and Y:",Sxy)
print(f"βx:{βx:.2f}")
print(f"βo:{βo:.2f}")
```

```python
#SSE_residuals
beta_x_reduced = βx
beta_0_reduced = βo


y_hat_reduced = [βo + βx * xi for xi in x]

residuals_reduced = [yi - yhi for yi, yhi in zip(y, y_hat_reduced)]
SSE_reduced = sum([r**2 for r in residuals_reduced])

print(f"SSE Reduced Model = {SSE_reduced:.2f}")
```

```python
#SSE_Ancova
y_hat_full = [beta0 + betaG * Gi + betaX * xi 
              for Gi, xi in zip(G, x)]

residuals_full = [yi - yhi for yi, yhi in zip(y, y_hat_full)]
SSE_ANCOVA = sum([r**2 for r in residuals_full])

print(f"SSE ANCOVA Model = {SSE_ANCOVA:.2f}")
```

```python
#Explained Variability SST of reduced Model
SSR_residual = total_y_minus_ymean_sq2 - SSE_reduced
print(f"SSR_residual = {SSR_residual:.2f}")
```

```python
#Explained Variability of Full Model
SSR_Full = total_y_minus_ymean_sq2 - SSE_ANCOVA
print(f"SSR_Full = {SSR_Full:.2f}")
```

```python
#Extra information explained by adding group variable
SSR_G_given_x = SSR_Full - SSR_residual
print(f"SSR_G|x = {SSR_G_given_x:.2f}")
```

```python
#F_statistic
parameters_of_ancova = 3
num_groups = len(set(G))
print("Number of groups =", num_groups)
Degree_of_group = num_groups - 1
Degree_of_freedom = len(x) - parameters_of_ancova

F_Statistic = (SSR_G_given_x / Degree_of_group) / (SSE_ANCOVA / Degree_of_freedom)
print(f"F_Statistic = {F_Statistic:.2f}")
```

```python
#F(1,5) = 32
#Look at F(1,5) in the F_Statistic Table at α = 0.05
#At F(1,5) at α = 0.05 = 6.61

#Therefore, if F_Calculated_Value > F_Theoritical_Value then Ho is rejected and Group information is significant

F_Calculated_Value = F_Statistic
F_Theoretical = 6.61  # Value from F-table

if F_Calculated_Value > F_Theoretical:
    print(f"H0 is rejected because {F_Calculated_Value:.2f} > {F_Theoretical}. \nThe group information is significant.")
else:
    print("H0 is not rejected.\nThe group information is NOT significant.")
```
