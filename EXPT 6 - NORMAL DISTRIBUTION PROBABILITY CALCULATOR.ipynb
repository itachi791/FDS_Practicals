```python
 import numpy as np
 from scipy.stats import norm
 
```

```python
data_input = input("Enter your data values separated by commas: ")
data = [float(value) for value in data_input.split(',')]

# Calculate mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data, ddof=1)

print(f"\nMean = {mean:.4f}")
print(f"Standard Deviation = {std_dev:.4f}")

# Get the value of x and condition
x = float(input("\nEnter the value of x: "))
condition = input("Do you want P(X > x) or P(X < x)? (enter '>' or '<'): ").strip()

#Probabikity
if condition == '>':
    probability = 1 - norm.cdf(x, loc=mean, scale=std_dev)
    print(f"\nP(X > {x}) = {probability:.6f}")
elif condition == '<':
    probability = norm.cdf(x, loc=mean, scale=std_dev)
    print(f"\nP(X < {x}) = {probability:.6f}")
else:
    print("\nInvalid input! Please enter '>' or '<'.")
```
